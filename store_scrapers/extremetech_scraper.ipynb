```python
import requests
from bs4 import BeautifulSoup
import json
from datetime import datetime

# Configuración de cabeceras para simular un navegador real
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# URL de prueba (Tarjetas de Video de ExtremeTech)
TEST_URL = "https://extremetechcr.com/product-category/componentes/tarjetas-de-video/"

print("Iniciando prueba de scraping en ExtremeTech...")

try:
    response = requests.get(TEST_URL, headers=HEADERS, timeout=10)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Intentamos buscar los contenedores de productos típicos
        # Nota: Si el HTML cambió, ajustaremos estas clases en el siguiente paso
        products = soup.find_all('article', class_='product-miniature')
        
        if not products:
            # Intento alternativo por si usan otra estructura de divs
            products = soup.find_all('div', class_='product-description')
            
        print(f"Se encontraron {len(products)} productos en la primera página.")
        
        # Imprimir una muestra del primer producto para analizar su estructura interna
        if products:
            print("\n--- Estructura del primer producto encontrado ---")
            print(products[0].prettify()[:1000]) # Mostramos los primeros 1000 caracteres
    else:
        print(f"Error al conectar: Código de estado {response.status_code}")

except Exception as e:
    print(f"Ocurrió un error en la petición: {str(e)}")
```
